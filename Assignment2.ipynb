## 1. Sample Data

```python

documents = [
    "I love machine learning",
    "machine learning is powerful",
    "I love data science",
    "data science uses machine learning"
]
```

## 2. Bag of Words (Count Occurrence)

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
bow = cv.fit_transform(documents)

print("Vocabulary:")
print(cv.get_feature_names_out())

print("\nBoW Matrix:")
print(bow.toarray())
```

## 3. Normalized Bag of Words

```python
import numpy as np

bow_array = bow.toarray()
bow_l1 = bow_array / bow_array.sum(axis=1, keepdims=True)

print("\nL1 Normalized BoW:")
print(bow_l1)
```

```python
from sklearn.preprocessing import normalize

bow_l2 = normalize(bow_array, norm='l2')

print("\nL2 Normalized BoW:")
print(bow_l2)
```

## 4. TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(documents)

print("\nTF-IDF Vocabulary:")
print(tfidf.get_feature_names_out())

print("\nTF-IDF Matrix:")
print(tfidf_matrix.toarray())
```

## 5. Word2Vec Embeddings

```python
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec

tokenized_docs = [word_tokenize(doc.lower()) for doc in documents]

w2v_model = Word2Vec(
    sentences=tokenized_docs,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4
)

print("\nWord2Vec embedding for 'machine':")
print(w2v_model.wv['machine'])

print("\nWords similar to 'machine':")
print(w2v_model.wv.most_similar('machine'))
```

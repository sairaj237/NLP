```python
import nltk
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
print("Ready!")
```

```python
target_word = "plant"

sentence_a = "The industrial plant buildings manufacture products."
sentence_b = "The living plant organism has green leaves and roots."

# Tokenize both sentences
tokens_a = word_tokenize(sentence_a)
tokens_b = word_tokenize(sentence_b)
```

```python
# Apply Lesk to find the correct synset (Synonym Set from WordNet)
# pos='n' tells the algorithm we are strictly looking for noun definitions
sense_a = lesk(tokens_a, target_word, pos='n')
sense_b = lesk(tokens_b, target_word, pos='n')

print(f"WordNet Synset for Context A: {sense_a}")
print(f"WordNet Synset for Context B: {sense_b}")
```

```python
print("--- AI Meaning Interpretation ---")

print(f"\nSentence A: '{sentence_a}'")
if sense_a:
    print(f"Selected Meaning: {sense_a.name()}")
    print(f"Definition: {sense_a.definition()}")
else:
    print("Could not disambiguate.")

print(f"\nSentence B: '{sentence_b}'")
if sense_b:
    print(f"Selected Meaning: {sense_b.name()}")
    print(f"Definition: {sense_b.definition()}")
else:
    print("Could not disambiguate.")
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = "NLTK is great! I'm learning NLP, tweeting #AI and loving it :)"
```

# 2. Tokenization

a) Whitespace Tokenization

```python
from nltk.tokenize import WhitespaceTokenizer

wt = WhitespaceTokenizer()
print(wt.tokenize(text))
```

b) Punctuation-based Tokenization

```python
from nltk.tokenize import WordPunctTokenizer

wpt = WordPunctTokenizer()
print(wpt.tokenize(text))
```

c) Treebank Tokenization

```python
from nltk.tokenize import TreebankWordTokenizer

tbt = TreebankWordTokenizer()
print(tbt.tokenize(text))
```

d) Tweet Tokenization

```python
from nltk.tokenize import TweetTokenizer

tt = TweetTokenizer()
print(tt.tokenize(text))
```

e) MWE (Multi-Word Expression) Tokenization

```python
from nltk.tokenize import MWETokenizer

mwe = MWETokenizer([('machine', 'learning'), ('artificial', 'intelligence')])
print(mwe.tokenize("I like machine learning and artificial intelligence".split()))
```

# 3. Stemming

a) Porter Stemmer

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
words = ['running', 'runs', 'ran', 'easily', 'fairly']
print([ps.stem(w) for w in words])
```

b) Snowball Stemmer

```python
from nltk.stem import SnowballStemmer

ss = SnowballStemmer("english")
print([ss.stem(w) for w in words])
```

# 4. Lemmatization (WordNet)

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
words = ['running', 'better', 'cars']

print([lemmatizer.lemmatize(w, pos='v') for w in words])
print([lemmatizer.lemmatize(w, pos='n') for w in words])
```

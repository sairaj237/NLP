Perform tokenization (Whitespace, Punctuation-based, Treebank, Tweet, MWE) using NLTK
library. Use porter stemmer and snowball stemmer for stemming. Use any technique for
lemmatization.

Import libraries

```python
import nltk
from nltk.tokenize import WhitespaceTokenizer, WordPunctTokenizer, TreebankWordTokenizer, TweetTokenizer, MWETokenizer
from nltk.stem import PorterStemmer, SnowballStemmer, WordNetLemmatizer

nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
```

Define sample text

```python
text = "NLTK is a leading platform for building Python programs to work with human language data. It provides easy-to-use interfaces to over 50 corpora and lexical resources such as WordNet, along with a suite of text processing libraries for classification, tokenization, stemming, tagging, parsing, and semantic reasoning, wrappers for industrial-strength NLP libraries, and an active discussion forum."
tweet = "@nltk Text processing is fun! #NLP :)"
mwe_text = "Standard Operating Procedure is to follow the rules."
```

Whitespace Tokenization

```python
tokenizer = WhitespaceTokenizer()
print(tokenizer.tokenize(text))
```

Punctuation-based Tokenization

```python
tokenizer = WordPunctTokenizer()
print(tokenizer.tokenize(text))
```

Treebank Tokenization

```python
tokenizer = TreebankWordTokenizer()
print(tokenizer.tokenize(text))
```

Tweet Tokenization

```python
tokenizer = TweetTokenizer()
print(tokenizer.tokenize(tweet))
```

MWE Tokenization

```python
tokenizer = MWETokenizer([('Standard', 'Operating', 'Procedure')])
print(tokenizer.tokenize(mwe_text.split()))
```

Porter Stemmer

```python
# words = tokenizer.tokenize(text)
words = ["running", "jumps", "easily", "fairly"]
```

```python
stemmer = PorterStemmer()
print([stemmer.stem(word) for word in words])
```

Snowball Stemmer

```python
stemmer = SnowballStemmer("english")
print([stemmer.stem(word) for word in words])
```

Lemmatization

```python
lemmatizer = WordNetLemmatizer()
print([lemmatizer.lemmatize(word) for word in words])
```
